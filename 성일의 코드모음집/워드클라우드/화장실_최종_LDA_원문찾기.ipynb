```python
import pickle
path = "화장실_최종_군집_5.pkl"

with open(f"{path}", "rb") as d:
    df = pickle.load(d)


data_f = df[(df['LDAvis_action_cluster'] == 2) | (df['LDAvis_action_cluster'] == 6) | (df['LDAvis_action_cluster'] == 8)]

for i in data_f['reply_og']:
    if "" in i:
        print(i)
        print("="*80)
```

```python
from collections import Counter

for_plot = []
for i in data_f["tagged_review"]:
    for j in i:
        for_plot.append(j)
data = Counter(for_plot)

# data['온수'] = 300
# data['온도'] = 250


from wordcloud import WordCloud
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rc('font', family='NanumGothicCoding')
mpl.rcParams['axes.unicode_minus'] = False

wc = WordCloud(background_color = 'white', 
            font_path="/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf",
            width=800,
            height=300)
wc.generate_from_frequencies(data)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
